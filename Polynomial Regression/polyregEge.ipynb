```python
data = pd.read_csv("linear.csv")
```

```python
x = data["metrekare"]
y = data["fiyat"]
```

```python
plt.scatter(x,y)
```

```python
x = np.array(x)
y = np.array(y)
```

```python
a,b,c,d = np.polyfit(x,y,3)
```

```python
z = np.arange(150)
```

```python
plt.scatter(x,y)
plt.plot(z,a*z*z*z+b*z*z+c*z+d, c="black")
```

```python
print(a, "x^3+",b, "x^2+",c,"x+",d)
```

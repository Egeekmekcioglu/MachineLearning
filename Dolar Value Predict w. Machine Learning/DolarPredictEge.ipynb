```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
```

```python
data = pd.read_csv("2016dolaralis.csv")
print(data)
```

```python
x = data["Gun"]
y = data["Fiyat"]
```

```python
x = x.values.reshape(251,1)
y = y.values.reshape(251,1)
```

```python
plt.scatter(x,y)
plt.show()
```

# Linear Regression

```python
Predictlinear = LinearRegression()               # Linear Regression ( Prediction )
Predictlinear.fit(x,y)
Predictlinear.predict(x)
plt.scatter(x,y)
plt.plot(x,Predictlinear.predict(x),c="red")
```

# 2. Degree Polynomial Regression

```python
PredictPolynom = PolynomialFeatures(degree=2)                 
x_new = PredictPolynom.fit_transform(x)

PolynomModel = LinearRegression()
PolynomModel.fit(x_new,y)
PolynomModel.predict(x_new)

plt.scatter(x,y)
plt.plot(x,PolynomModel.predict(x_new),c="midnightblue")
```

# 3. Degree Polynomial Regression

```python
PredictPolynomx = PolynomialFeatures(degree=3)                 
x_new = PredictPolynomx.fit_transform(x)

PolynomModelx = LinearRegression()
PolynomModelx.fit(x_new,y)
PolynomModelx.predict(x_new)

plt.scatter(x,y)
plt.plot(x,PolynomModelx.predict(x_new),c="navy")
```

# Error Square Function for which is better 

```python
ErrorSqlinear = 0                     # 3.Degree Regression needed to ignore before running.
ErrorSqpoly = 0

for i in range(len(x_new)):
    ErrorSqpoly = ErrorSqpoly + (float(y[i])-float(PolynomModel.predict(x_new)[i]))**2
    
for i in range(len(y)):
    ErrorSqlinear = ErrorSqlinear + (float(y[i])-float(Predictlinear.predict(x)[i]))**2
    
```

```python
print("Error Square 2.Deg Polynom: ",ErrorSqpoly)
print("Error Square Linear: ",ErrorSqlinear)
```

```python
ErrorSqpoly = 0

for a in range(150):
    PredictPolynom = PolynomialFeatures(degree=a+1)                 
    x_new = PredictPolynom.fit_transform(x)

    PolynomModel = LinearRegression()
    PolynomModel.fit(x_new,y)
    PolynomModel.predict(x_new)

    for i in range(len(x_new)):
        ErrorSqpoly = ErrorSqpoly + (float(y[i])-float(PolynomModel.predict(x_new)[i]))**2
    print(a+1,".Degree Polynom Error ",ErrorSqpoly)
    ErrorSqpoly=0
```

# Better Result for 8.Degree Regression

```python
PredictPolynom8 = PolynomialFeatures(degree=8)                 
x_new = PredictPolynom8.fit_transform(x)

PolynomModel8 = LinearRegression()
PolynomModel8.fit(x_new,y)
PolynomModel8.predict(x_new)

plt.scatter(x,y)
plt.plot(x,PolynomModel8.predict(x_new),c="indigo")      # Error Func : 8 .Degree Polynom Error  0.3815800293210352
```

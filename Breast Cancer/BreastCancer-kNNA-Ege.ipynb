```python
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
data = pd.read_csv("breast-cancer-wisconsin.data")
```

```python
data.replace('?', -99999, inplace=True)
data.drop(['id'], axis=1)
```

```python
y = data.benormal
x = data.drop(['benormal'], axis=1)
```

```python
imp = SimpleImputer(missing_values= np.NAN, strategy= 'mean', fill_value=None, verbose=0, copy=True)

x = imp.fit_transform(x)
```

```python
predict = KNeighborsClassifier(n_neighbors=3, weights='uniform', algorithm='auto', leaf_size=30, p=2, metric='euclidean', metric_params=None, n_jobs=1)
predict.fit(x,y)
ypredict = predict.predict(x)
```

```python
success = accuracy_score(y, ypredict, normalize=True, sample_weight=None)
print("Success ",success*100," in Percent:" )

print(predict.predict([1,2,2,2,3,2,1,2,3,2]))
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression as lr
from sklearn.preprocessing import PolynomialFeatures as pf
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
from sklearn import linear_model
```

```python
data = pd.read_csv("linear.csv")
```

```python
x = data["metrekare"]
y = data["fiyat"]
plt.scatter(x,y)
```

```python
m, b = np.polyfit(x,y,1)
a = np.arange(150)
plt.scatter(x,y)
plt.plot(m*a+b,c="black")
```

```python
plt.show()
print("y=",m,"x+",b)
```

# Difference Square = Ridge Regression
# Difference Absolute Value = Lasso Regression

```python
for i in range(len(y)):
    ErrorSq = (y[i]-(m*x[i]+b))**2          # Ridge Regression

print(ErrorSq)                               # If we did the same operation with the polynomial regression. We would reach the value of circa 111.
                                            # Also in short, it produced a better result.(Polynomial better in same story)
```

```python
EPU = ErrorSq/(len(y))      # Error per unit.
print(EPU)
```
